```python
nation_deposit = df.groupby("nationality")["bank_deposits"].mean().round(2)
```

```python
nation_deposit.sort_values().plot(kind='barh', color='teal')
plt.title("Average Bank Deposits by Nationality")
plt.xlabel("Average Deposit")
plt.ylabel("Nationality")
#plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
nation_deposit_1 = df.groupby("nationality")["checking_accounts"].mean().round(2)
```

```python
nation_deposit_1.sort_values().plot(kind='barh', color = ["coral", "teal",])
plt.title("Average checking Accounts by Nationality")
plt.xlabel("Checking Account")
plt.ylabel("Nationality")
plt.tight_layout()
plt.show()
```

```python
nation_deposit_2 = df.groupby("nationality")["saving_accounts"].mean().round(2)
```

```python
nation_deposit_2.sort_values().plot(kind='bar', color = ["Aqua", "teal",])
plt.title("Average Saving Accounts by Nationality")
plt.xlabel("Saving Account")
plt.ylabel("Nationality")
plt.tight_layout()
plt.show()
```

```python
nation_deposit3 = df.groupby("nationality")["foreign_currency_account"].mean().round(2)
```

```python
nation_deposit3.sort_values().plot(kind='barh', color = ['#b46c6c', '#944242', '#6c2a2a', '#3f1a1a'])
plt.title("Average Foreign Accounts by Nationality")
plt.xlabel("Foregin Account")
plt.ylabel("Nationality")
plt.tight_layout()
plt.show()
```

```python
nation_deposit4 = df.groupby("nationality")["business_lending"].mean().round(2)
```

```python
nation_deposit4.sort_values().plot(kind='barh', color = ["teal", "grey",])
plt.title("Average Lending by Nationality")
plt.xlabel("Lending")
plt.ylabel("Nationality")
plt.tight_layout()

plt.show()
```

```python
prop_occ = df.groupby("nationality")["properties_owned"].size()
```

```python
plt.figure(figsize=(6, 6)) 
prop_occ.sort_values().plot(kind='pie', labels= prop_occ.index, autopct="%1.1f%%", color='#1f585c') 
plt.title("Properties by Nationality") 
plt.tight_layout() 
plt.legend(loc="best")
plt.show()
```

```python
numerical_col = ['Estimated Income', 'Superannuation Savings']

plt.figure(figsize=(12, 5))
for i, col in enumerate(numerical_col):
    plt.subplot(1, 2, i + 1)
    sns.histplot(df[col], kde=True, color='#1f585c', edgecolor='white')
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
numerical_col = ['Saving Accounts','Bank Loans']

plt.figure(figsize=(12, 5))
for i, col in enumerate(numerical_col):
    plt.subplot(1, 2, i + 1)
    sns.histplot(df[col], kde=True, color='teal', edgecolor='white')
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
